```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
imgs_path = "../Pokemon Dataset/pokemon_jpg/pokemon_jpg/*.jpg"
imgs_path = glob.glob(imgs_path)
```

```python
index = 1
for img in imgs_path:
    pokemon = cv2.imread(img)
    edges_detection = cv2.Canny(pokemon,100,200)
    image = [pokemon, edges_detection]
    cv2.imwrite('./edge_img/' + str(index) + '.jpg', edges_detection)
    index+=1
```

```python
location = [121, 122] 
for loc, edge_image in zip(location, image): 
   plt.subplot(loc) 
   plt.imshow(edge_image, cmap='gray')
```
